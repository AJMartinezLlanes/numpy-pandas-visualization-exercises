```python
import matplotlib.pyplot as plt
import math
print(plt)
print(math)
```

```python
x=list(range(1, 20))
y1=[0*n+1 for n in x]
y2=[math.log(n) for n in x]
y3=[n for n in x]
y4=[n*math.log(n) for n in x]
y5=[n**2 for n in x]
y6=[2**n for n in x]
y7=[math.factorial(n) for n in x]
y8=[n**n for n in x]

plt.plot(x, y1, c='green', label='O(1)')
plt.plot(x, y2, c='red', label='O(log n)')
plt.plot(x, y3, c='orange', label='O(n)')
plt.plot(x, y4, c='pink', label='O(n log n)')
plt.plot(x, y5, c='blue', label='O(n^2)')
plt.plot(x, y6, c='yellow', label='O(2^n)')
plt.plot(x, y7, c='purple', label='O(n!)')
plt.plot(x, y8, c='brown', label='O(n^n)')

plt.legend(loc='upper right')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.ylim(0,30)
plt.xlim(0,20)
plt.show()
```
