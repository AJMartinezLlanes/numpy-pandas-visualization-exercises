```python
import matplotlib.pyplot as plt
```

1. Use matplotlib to plot the following equation:

    * y = x^2 − x + 2

    You'll need to write the code that generates the x and y points.

    Add an anotation for the point 0, 0, the origin.



```python
x = list(range(-50, 50))
y = [n ** 2 - n + 2 for n in x]
plt.plot(x, y)
plt.text(1, 0, '(0, 0)', fontsize=10, color='blue')
plt.show()
```

```python
plt.title('y=x^3')
x = list(range(-50, 50))
y = [n**3 for n in x]
plt.plot(x, y)
plt.text(0.25, 0, '(0, 0)', fontsize=10, color='blue')
plt.show()
```

```python
plt.title('y=2^x')
x = list(range(-5,50))
y = [2**n for n in x]
plt.plot(x, y)
plt.text(0.25, 0, '(0, 0)', fontsize=10, color='blue')
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
# plot y=√x
plt.subplot(2,2,1)
plt.title('y=√x')
x = list(range(0, 50))
y = [math.sqrt(n) for n in x]
plt.plot(x, y, c='green')
plt.text(0.25, 0, '(0, 0)', fontsize=10, color='blue')
plt.show()

# plot y=x^3
plt.subplot(2,2,2)
plt.title('y=x^3')
x = list(range(-50, 50))
y = [n**3 for n in x]
plt.plot(x, y, c='red')
plt.text(0.25, 0, '(0, 0)', fontsize=10, color='blue')
plt.show()

# plot y=2^x
plt.subplot(2,2,3)
plt.title('y=2^x')
x = list(range(0, 50))
y = [2**n for n in x]
plt.plot(x, y, c='orange')
plt.text(0.25, 0, '(0, 0)', fontsize=10, color='blue')
plt.show()

# plot y=1/n+1
plt.subplot(2,2,4)
plt.title('y=1/n+1')
x = list(range(1, 50))
y = [1/n+1 for n in x]
plt.plot(x, y, c='pink')
x = list(range(-50, -2))
y = [1/n+1 for n in x]
plt.plot(x, y, c='cyan')
plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = list(range(1, 50))
# plot1 y=√x
y1 = [math.sqrt(n) for n in x]
# plot2 y=x^3
y2 = [n**3 for n in x]
# plot3 y=2^x
y3 = [2**n for n in x]
# plot4 y=1/n+1
y4 = [1/n+1 for n in x]

plt.plot(x, y1, c='green', label='y=√x')
plt.plot(x, y2, c='red', label='y=x^3')
plt.plot(x, y3, c='orange', label='y=2^x')
plt.plot(x, y4, c='pink', label='y=1/n+1')
plt.legend(loc='upper left')
plt.title('Combination of Figures')
plt.ylim(-5,50)
plt.xlim(-5,50)
plt.show()
```
